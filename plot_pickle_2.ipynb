```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'cmr10'
plt.rcParams['axes.formatter.use_mathtext'] = True
plt.rcParams['font.size'] = 12
```

```python
cv_scores = np.load('volumes_cv.npz')
f1_scores = cv_scores['f1_scores']
auc_scores = cv_scores['auc_scores']
```

```python
plt.figure(figsize=(4, 3), dpi=300)
plt.plot(range(1, len(auc_scores) + 1), auc_scores, color='b', linestyle='-', label=f'ROC AUC per fold')
plt.axhline(y=np.mean(auc_scores), color='b', linestyle='--', label=f'Mean AUC ({np.mean(auc_scores):.2f}, $\\sigma$={np.std(auc_scores):.2f})')
plt.plot(range(1, len(f1_scores) + 1), f1_scores, color='r', linestyle='-', label=f'f1 per fold')
plt.axhline(y=np.mean(f1_scores), color='r', linestyle='--', label=f'Mean f1 ({np.mean(f1_scores):.2f}, $\\sigma$={np.std(f1_scores):.2f})')
plt.xlabel('Fold Number')
plt.ylabel(f'Score')
plt.xticks(ticks=range(1, 11), labels=range(1, 11))
plt.ylim(0, 1)
plt.legend(loc='lower left')
plt.grid(True)
plt.tight_layout()
plt.savefig('figures/volumes_cv.png', dpi=300, bbox_inches='tight')
plt.show()
```
